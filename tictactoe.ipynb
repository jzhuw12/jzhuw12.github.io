```python
from IPython.display import clear_output
import random

def display_board(board):

    print(' ' + board[1] + ' | '+ board[2] + ' | '+ board[3]+ ' ')
    print(' --------- ')
    print(' ' + board[4] + ' | '+ board[5] + ' | '+ board[6]+ ' ')
    print(' --------- ')
    print(' ' + board[7] + ' | '+ board[8] + ' | '+ board[9]+ ' ')


def player_input():
    
    marker1 = ''
    
    while marker1 != 'X' and marker1 != 'O':
        marker1 = input('Please choose a marker for 1st player,X or O: ').upper()
        
        if marker1 == 'X':
            marker2 = 'O'
        else:
            marker2 = 'X'
    print('Marker for player1 is: ' + marker1 + ', Marker for player2 is: ' + marker2)
    
    return(marker1,marker2)
            

def place_marker(board, marker, position):
    
    board[position] = marker

    
def win_check(board, mark):

    return ((board[7] == mark and board[8] == mark and board[9] == mark) or # across the top
    (board[4] == mark and board[5] == mark and board[6] == mark) or # across the middle
    (board[1] == mark and board[2] == mark and board[3] == mark) or # across the bottom
    (board[7] == mark and board[4] == mark and board[1] == mark) or # down the middle
    (board[8] == mark and board[5] == mark and board[2] == mark) or # down the middle
    (board[9] == mark and board[6] == mark and board[3] == mark) or # down the right side
    (board[7] == mark and board[5] == mark and board[3] == mark) or # diagonal
    (board[9] == mark and board[5] == mark and board[1] == mark)) # diagonal


def choose_first():
    num = random.randint(1,2)
    if num == 1:
        return 'player1'
    else:
        return 'player2'
    
    
def space_check(board, position):
    
    return board[position] == ' '

def full_board_check(board):
    
    for i in range(1,10):
        if space_check(board,i):
            return False
    return True


def player_choice(board,player):
    
    choice = 0
    
    while choice not in [1,2,3,4,5,6,7,8,9] or not space_check(board,choice):
        choice = int(input('{},please choose a position:(1-9)'.format(player)))

    return choice


def replay():
    
    play_again = 0
    
    while play_again != 'Y' and play_again != 'N':
        play_again = input('Do you want to play again? Reply Y or N: ')
        
    return play_again == 'Y'


    
print('Welcome to Tic Tac Toe!')

start = input('Press Q to start ')
print('\n')

while start == 'Q':

    board = [' ',' ',' ',' ',' ',' ',' ',' ',' ',' ']

    player1_marker, player2_marker = player_input()
    
    first = 0
    
    if choose_first() == 'player1':
        
        print('It\'s now ' + player1_marker + ' \'s turn.')
        
        first = player1_marker
        second = player2_marker
        
    else:
        
        print('It\'s now ' + player2_marker + ' \'s turn.')
        first = player2_marker
        second = player1_marker
            
    
    while (not win_check(board, 'X') and not win_check(board, 'O')) and not full_board_check(board):
        
        position = player_choice(board,first)
        place_marker(board, first, position)
        clear_output()
        display_board(board)
        
        if (not win_check(board, 'X') and not win_check(board, 'O')) and not full_board_check(board):
            
            position = player_choice(board,second)
            place_marker(board, second, position)
            clear_output()
            display_board(board)
            
            
    if win_check(board, player1_marker):
        
        print('\n')
        print('❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀')
        print('You win ! player one ({}). Good one.'.format(player1_marker))
        print('❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀')
        
    elif win_check(board, player2_marker):
        
        print('\n')
        print('❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀')
        print('You win ! player two ({}). Good one.'.format(player2_marker))
        print('❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀♫❀')

    print('\n')
    
    if not replay():
        
        print('\n')     
        start = input('Thank you for playing! Press Q to start again at any time. ')
        clear_output()

    
```
